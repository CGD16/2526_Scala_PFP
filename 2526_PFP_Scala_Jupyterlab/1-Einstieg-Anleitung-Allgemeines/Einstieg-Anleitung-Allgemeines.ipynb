#### Ersatz für die Scala **Read-Eval-Print-Loop (REPL)** bzw. **Scala-Dateien**:

Wenn Sie aufgefordert werden, die Scala _Read-Eval-Print-Loop (REPL)_ zu verwenden oder Code in einer Datei zu implementieren, dann geben Sie stattdessen den Scala-Ausdruck in das jeweils folgende Eingabefeld (sogenannte _"Zelle"_) ein und drücken Sie dann oben die "Play-Taste ▶" oder `Shift+Enter`, um ihn von Scala auswerten zu lassen.
   - Eine solche "REPL-Eingabezelle" erkennen Sie am grauen Hintergrund und dem `[ ]:` davor.
   - Die Auswertung selbst kann etwas dauern... Dass sie in Arbeit ist, erkennen Sie an `[*]:` vor der Zelle.
   - Nach der Auswertung erscheint stattdessen eine Nummer vor der Zelle und das Ergebnis steht direkt darunter.

Probieren Sie es gleich hier aus (z.B. "`1+2`" ohne Anführungszeichen eingeben):

#### Aufeinander **aufbauende Teilaufgaben** und **Testfälle** (kleine Schnelltests):

Inbesondere in Klausuren bauen die Teilaufgaben oft aufeinander auf. Zum Beispiel kann/soll eine in Teilaufgabe b) zu definierende Funktion eine vorgegebene oder in a) zu implementierende Funktion wiederverwenden.
   - Die "REPL-Zellen" tragen in der **Reihenfolge ihrer Ausführung** zum _"globalen Zustand"_ bei, solange der laufende Scala-Kernel nicht zurückgesetzt wird.
   - Sie können auch **mehrere Zellen** zur Ausführung senden, **ohne** auf die einzelnen Ergebnisse zu **warten**.
   - Es ist auch möglich, nur einzelne Zellen erneut auszuführen, z.B. nach einer "Korrektur weiter oben".

Um Sie nicht mit langen Testfällen abzulenken, sind die zugehörigen Felder ggf. **verborgen** ("collapsed code").
   - Sie werden **auch dann ausgeführt**, wenn sie **verborgen** sind, also nicht ganz angezeigt werden.
   - Sie können den Code anzeigen (auf den Code klicken) und wieder verbergen (Menü "View" ➔ "Collapse Selected Code").

Probieren Sie es gleich hier aus (klicken Sie in die erste Zelle und dann **3x schnell hintereinander** die "Play-Taste ▶" oben):

````scala
// Diese Zelle definiert eine Funktion, die in der nächsten Zelle verwendet wird:
def inc: Int => Int = _ + 1
````

````scala
// Diese Zelle verwendet die Funktion aus der vorangehenden Zelle:
def incTwice: Int => Int = n => inc(inc(n))
println(incTwice(42))
````

````scala
// Beispiel und Schnelltest:
// Diese Zelle "testet" beide vorangehend definierten Funktionen
// und kann daher erst danach ausgeführt werden:
assert(inc(42) == 43); print("✅")
assert(incTwice(42) == 44); print("✅")
````

<div class="alert alert-block alert-warning">
Hinweis: Falls Sie doch mal im Menü "Kernel" ➔ "Restart Kernel and Clear Outputs of All Cells..." gewählt haben, müssen Sie <b>alle benötigten Zellen nochmal ausführen</b>. Ggf. hilft hier das Menü "Kernel" ➔ "Restart Kernel and Run up to Selected Cell...", um alle benötigten Vorarbeiten in einem bereitzustellen.
</div>

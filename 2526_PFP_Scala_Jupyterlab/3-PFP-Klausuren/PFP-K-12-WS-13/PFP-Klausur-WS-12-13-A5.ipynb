Das **SnakeSort**-Verfahren sortiert eine zweidimensionale Matrix nach folgendem Verfahren: Zuerst werden im Schritt $i=0$ die Werte zeilenweise abwechselnd auf- bzw. absteigend sortiert. Dann werden im Schritt $i=1$ die Werte spaltenweise aufsteigend sortiert. Beide Schritte werden abwechselnd solange wiederholt, bis sich nach einem Schritt keine Änderung an der Matrix mehr ergibt.

**Beispiel:**  
$
	\begin{array}{c}
		\left[ \begin{array}{ccc} 6 & 3 & 7 \\ 1 & 9 & 5 \\ 2 & 8 & 4 \end{array} \right] \\ m_{0}
	\end{array}
	\xrightarrow[i=0]{zeilenweise}
	\begin{array}{c}
		\left[ \begin{array}{ccc} 3 & 6 & 7 \\ 9 & 5 & 1 \\ 2 & 4 & 8 \end{array} \right] \\ m_{1}
	\end{array}
	\xrightarrow[i=1]{spaltenweise}
	\begin{array}{c}
		\left[ \begin{array}{ccc} 2 & 4 & 1 \\ 3 & 5 & 7 \\ 9 & 6 & 8 \end{array} \right] \\ m_{2}
	\end{array}
	\xrightarrow[i=2]{zeilenweise}
	\begin{array}{c}
		\left[ \begin{array}{ccc} 1 & 2 & 4 \\ 7 & 5 & 3 \\ 6 & 8 & 9 \end{array} \right] \\ m_{3}
	\end{array}
$  
$
	\xrightarrow[i=3]{spaltenweise}
	\begin{array}{c}
		\left[ \begin{array}{ccc} 1 & 2 & 3 \\ 6 & 5 & 4 \\ 7 & 8 & 9 \end{array} \right] \\ m_{4}
	\end{array}
	\xrightarrow[i=4]{zeilenweise}
	\begin{array}{c}
		\left[ \begin{array}{ccc} 1 & 2 & 3 \\ 6 & 5 & 4 \\ 7 & 8 & 9 \end{array} \right] \\ m_{5}
	\end{array}
	\Rightarrow \text{\small{fertig, da keine Änderung in Schritt $i=4$}}
$

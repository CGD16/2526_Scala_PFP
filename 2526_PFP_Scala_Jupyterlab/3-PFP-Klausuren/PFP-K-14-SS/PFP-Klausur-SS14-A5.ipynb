Ein gerichteter Graph `G` ist als Liste von Adjazenzen `A(v, out)` gegeben. `v` ist dabei ein Knoten der Form `V(id)`. `out` ist die Liste der direkt von `v` erreichbaren Nachbarknoten:

````scala
case class V(id: Int)
case class A(v: V, out: List[V])
type G = List[A]
````

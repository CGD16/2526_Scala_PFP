Beim _Game of Life_ sind Zellen auf einem begrenzten rechteckigen Feld so angeordnet, dass an jeder Position $(x, y)$ des Feldes immer genau ein `Cell`-Objekt ist:

```scala
case class Cell(x: Int, y: Int, alive: Boolean)
type Field = List[Cell]
```

Zellen im Inneren des Feldes haben jeweils $8$ Nachbarn, Zellen am Rand haben weniger.  
Alle Zellen ändern ihren Zustand gleichzeitig per `tick` nach folgenden Regeln:  
- Eine Zelle mit $n < 2$ oder $n > 3$ lebenden Nachbarn (`alive`) ist nach dem `tick` tot.
- Eine Zelle mit $n = 3$ lebenden Nachbarn ist nach dem `tick` lebendig.
- In allen anderen Fällen behält eine Zelle ihren Zustand bei.

**Beispiel:** Eine Zelle ist `alive` (◼) oder `!alive` (□).

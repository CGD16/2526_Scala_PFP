<summary><i><span style='background:#b2ebf2'>Hinweis 1: Nachrichten an 0 - inklusive erste</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-info">

Laut Aufgabenstellung soll an jedem Index in der Ergebnisliste die Liste der zugehörigen Nachrichten stehen. D.h. der erste Eintrag (mit Index `0`) in der Rückgabe enthält **alle** Nachrichten, die an Empfänger `0` adressiert sind (wobei einzelne Teillisten auch leer sein können). Die Struktur des Gesamtergebnisses sieht also wir folgt aus:
- `LazyList(LazyList(<alle Nachrichten an 0>), LazyList(<alle Nachrichten an 1>), LazyList(<alle Nachrichten an 2>), ...)`

Beginnen wir hier mit der ersten Teilliste: `LazyList(<alle Nachrichten an 0>)`

<hr>

Das Muster **links** im **ersten** `case` trifft zu, wenn das Kopfelement der Eingabeliste an Empfänger `0` gerichtet ist. Daher muss dessen Nachricht `c` als erstes Element der ersten Ergebnisliste übernommen werden (**rechts** von `=>`).

In der Restliste `r` könnten aber noch mehr Nachrichten an Empfänger `0` adressiert sein. Daher muss `r` ebenfalls noch rekursiv mit `distribute` verarbeitet werden. Die Rückgabe des rekursiven Aufrufs ist eine `LazyList` der Form:
- `LazyList(LazyList(<alle Nachrichten NUR AUS r an 0>), LazyList(<alle Nachrichten NUR AUS r an 1>), ...)`

Demnach muss die obige Nachricht `c` noch am Anfang der rekursiv ermittelten `LazyList(<alle Nachrichten NUR AUS r an 0>)` ergänzt werden:
- `LazyList(c, <alle weiteren Nachrichten NUR AUS r an 0>)`

Wenn das Kopfelement keine Nachricht an Empfänger `0` enhält, dann trifft das **zweite** Muster (**zweites** `case`) zu. D.h. die Nachricht `m` ist an Empfänger $x > 0$ adressiert und gehört daher vorerst zurückgestellt, weil wir zuerst noch alle anderen Nachrichten an Empfänger `0` aus `r` extrahieren müssen.

Dazu rufen wir rekursiv `distribute` mit der Restliste `r` auf. Die Rückgabe dieses rekursiven Aufrufs ist eine LazyList der Form:

Damit enthält der erste Eintrag (mit Index `0`) in dieser Rückgabe **alle** Nachrichten, die an Empfänger `0` adressiert sind, da `m` ja garantiert **nicht** dazugehört. Daher wird dieses Kopfelement des Ergebnisses aus der Rekursion zum **Kopfelement des Endergebnisses**.

  - `case class Vertex[T](l:Tree[T], v:T, r:Tree[T]) extends Tree[T]`.
- _"Muster dürfen Mustervariablen nicht mehrfach verwenden"_, da der Compiler sonst den Fehler _"duplicate pattern variable"_ meldet!
- Eines der zentralen Konzepte funktionaler Programmierung ist das sogenannte _"Currying"_, bei dem Funktionen _"unterversorgt"_ bleiben dürfen.
- Parallelisierung kommt (fast) immer mit zusätzlichem _"Overhead"_ (Verwaltung der Threads, Synchronisation, ...) daher.

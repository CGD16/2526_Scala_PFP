"Regel 60" ist ein einfacher zellulärer Automat. Dieser besteht aus einem unendlichen, eindimensionalen Band aus Zellen, die entweder aktiv (◼) oder inaktiv (□) sein können. Dieses Band verändert sich schrittweise, wobei in jedem Schritt für jede Zelle der eigene Zustand sowie der des linken Nachbarn vor dem Schritt betrachtet wird. Die zeitliche Entwicklung ausgehend von einer aktiven Zelle ist rechts beispielhaft gezeigt.

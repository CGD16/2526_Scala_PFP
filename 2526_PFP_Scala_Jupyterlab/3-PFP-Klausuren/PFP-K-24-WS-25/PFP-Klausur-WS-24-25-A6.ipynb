<summary><i><span style='background:#b2ebf2'>Hinweis</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-info">

Idee: Traversiere die Tupel $(f,t)$ der Entschlüsselungstabelle **rekursiv** bis
- entweder die komplette Entschlüsselungstabelle erfolglos abgearbeitet ist => dann soll das Eingabezeichen _"unverändert zurückgegeben werden"_;
- oder das gesuchte Zeichen als erstes Tupelelement $f$ vorkommt => dann ist das zweite Tupelelement $t$ der gesuchte Rückgabewert.

Die Besonderheit hier ist, dass `translateChar` eine zweistellige Funktion ist (vom Typ: _"Entschlüsselungstabelle => Eingabezeichen => Ausgabezeichen"_), das vorgedruckte _Pattern Matching_ aber direkt auf das erste Argument (die Entschlüsselungstabelle selbst) angewandt werden soll. D.h. das zu entschlüsselnde Eingabezeichen muss durch eine Funktion _"konsumiert"_ werden, die erst noch auf der **rechten Seite** der einzelnen `case`-Zeilen definiert und zurückgegeben werden muss!

Für den Basisfall im ersten `case` ergibt sich damit z.B.: `case Nil => (x => x)`. Dabei ist `x => x` eine **Funktion**, die das Eingabezeichen unverändert zurückgibt.

Der Rekursionsfall zerlegt zunächst die Entschlüsselungstabelle z.B. mittels: `case (f,t) :: remDict => ...`.  
Die Rückgabe auf der **rechten Seite** muss nun wiederum eine **Funktion** vom Typ `Char => Char` sein. Hier bietet sich erneut _Pattern Matching_ an, um zu unterscheiden, ob `f` das gesuchte Eingabezeichen ist - oder eine Rekursion auf die restliche Entschlüsselungstabelle `remDict` notwendig ist.

</div>
</details>

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

```scala
def translateChar: List[(Char,Char)] => Char => Char = {
   case Nil => (x => x)
   case (f,t) :: remDict => {
      case c if c == f => t
      case c => translateChar(remDict)(c)
   }
}  
```

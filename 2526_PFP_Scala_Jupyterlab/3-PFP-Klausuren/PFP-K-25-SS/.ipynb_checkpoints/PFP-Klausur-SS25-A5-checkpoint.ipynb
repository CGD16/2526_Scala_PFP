# Aufgabe 5 (Postsches Korrespondenzproblem)

_Verwenden Sie ausschließlich aus der Lehrveranstaltung bekannte Scala-Konstrukte, dabei sind `var` und seiteneffektbehaftete Methoden/Klassen/Objekte strikt untersagt._

Gegeben ist eine Folge an Dominosteinen, auf deren Ober- und Unterseite jeweils eine Zeichenfolge steht. Die Nummer eines Dominosteins ist seine Position in dieser Folge. Eine Lösung des Postschen Korrespondenzproblems ist eine Liste beliebig vieler Steine, sodass die Konkatenation der Zeichenfolgen der oberen Seiten der Steine die gleiche Zeichenfolge ergibt, wie die der unteren Seiten. Ein Dominostein darf dabei beliebig oft verwendet werden. Die Lösungsliste ist die Liste der Dominosteinnummern.

```scala
type Dominoes[A] = List[(List[A], List[A])] // Liste der Steine als Paar "(Folge oben, Folge unten)"
def solve[A] : Dominoes[A] => List[Int] =   
   dominoes => solveRec(dominoes)(Nil)(Nil).reverse 
```

**b)** Vervollständigen Sie die Funktion `candidates`. Sie erhält die gegebene Liste von Dominosteinen sowie eine partielle Lösung, bestehend aus der oberen (`topS`) und unteren (`botS`) konkatenierten Zeichenfolge. Sie sollen diese partielle Lösung mit jedem Dominostein erweitern und nur solche Kombinationen herausfiltern, deren konkatenierten Zeichenfolgen Präfixe voneinander sind. Die resultierenden Kandidaten werden als 3er-Tupel zurückgeben (konkatenierte Zeichenfolge oben, konkatenierte Zeichenfolge unten, Nummer des Steins).

````scala
type Dominoes[A] = List[(List[A], List[A])] // Liste der Steine als Paar "(Folge oben, Folge unten)"
````

````scala
def candidates[A] : Dominoes[A] => List[A] => List[A]
   => List[(List[A], List[A], Int)] = dominoes => topS => botS =>
      dominoes. /*** IHR CODE HIER (Hinweis 1) ***/ // Nummerieren
         .map({case ((a, b), i) => ( /*** IHR CODE HIER (Hinweis 2) ***/ ,
            /* Woerter anhaengen */ /*** IHR CODE HIER (Hinweis 3) ***/ , i)})
         . /*** IHR CODE HIER (Hinweis 4) ***/ // Praefixe suchen
````

````scala
// Beispiel und Schnelltest:
assert(candidates(Nil)(Nil)(Nil) == Nil); print("✅")
assert(candidates(List((List('a'),List('a','b')), (List('a','b'),List('y', 'z')), (List('y','z'),List('y'))))
                 (Nil)(Nil)
       == List((List('a'),List('a','b'),0), (List('y','z'),List('y'),2))); print("✅")
assert(candidates(List((List('a'),List('a','b')), (List('a','b'),List('y', 'z')), (List('y','z'),List('y'))))
                 (List('a','a'))(List('a','a','a'))
       == List((List('a','a','a'),List('a','a','a','a','b'),0))); print("✅")
assert(candidates(List((List('a'),List('a','b')), (List('a','b'),List('y', 'z')), (List('y','z'),List('y'))))
                 (List('x','x'))(List('x','x'))
       == List((List('x','x','a'),List('x','x','a','b'),0), (List('x','x','y','z'),List('x','x','y'),2))); print("✅")
assert(candidates(List((List('a'),List('a','b','a')), (List('a','b'),List('b','b')), (List('b','a','a'),List('a','a'))))
                 (Nil)(Nil)
       == List((List('a'),List('a','b','a'),0))); print("✅")
````

**c)** Vervollständigen Sie den Backtracking-Algorithmus. `solveRec` erhält die Liste der Dominosteine, sowie die derzeitige obere und untere konkatenierte Zeichenfolge. Probieren Sie pro Rekursionsschritt alle Erweiterungskandidaten der Folgen aus (Teilaufgabe (b)). Sind zwei Kandidaten gleich, stellt die Zeichenfolge eine Lösung für das Korrespondenzproblem dar. Die erste gefundene Lösung soll zurückgegeben werden.

````scala
def solveRec[A] : Dominoes[A] => List[A] => List[A] => List[Int] =
   dominoes => topS => botS => // Kandidaten ermitteln:
      /*** IHR CODE HIER (Hinweis 1) ***/
         .foldLeft(Nil : List[Int])({
            case (Nil, (a, b, i)) if a == b => List(i) // Basisfall
            case (Nil, (a, b, i)) => // Rekursion:
               /*** IHR CODE HIER (Hinweis 2) ***/ match {
                  case Nil => Nil // keine Loesung mit Rekursion
                  case /*** IHR CODE HIER (Hinweis 3) ***/ // Loesung gefunden
               }
            case (sol, _) => sol // gefundene Loesung zurueckgeben
  })
````

````scala
def solve[A] : Dominoes[A] => List[Int] =
   words => solveRec(words)(Nil)(Nil).reverse
````

````scala
// Beispiel und Schnelltest:
assert(solve(List(("a".toList,"baa".toList),("ab".toList,"aa".toList),("bba".toList,"bb".toList) )) == List(0,2,1,2)); print("✅")
````

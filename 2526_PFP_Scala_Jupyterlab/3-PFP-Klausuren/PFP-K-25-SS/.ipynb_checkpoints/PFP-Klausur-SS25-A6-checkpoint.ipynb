# Aufgabe 6 (Nonogramme)

_Verwenden Sie ausschließlich aus der Lehrveranstaltung bekannte Scala-Konstrukte, dabei sind `var` und seiteneffektbehaftete Methoden/Klassen/Objekte strikt untersagt._

Ein Nonogramm ist ein Rätsel. Gegeben ist dazu ein rechteckiges Raster von zunächst weißen Zellen, sowie eine Spezifikation, die für jede Spalte und jede Zeile die Länge der zusammenhängenden schwarzen Bereiche (die durch mindestens eine weiße Zelle von einander getrennt sind) in der Lösung beschreibt. In der nebenstehenden Abbildung spezifiziert z.B. die 2-elementige Liste (1,2) für die erste Zeile des Rasters, dass eine Lösung 2 schwarze Bereiche mit den Längen 1 bzw. 2 haben muss. Die Abbildung rechts zeigt eine Lösung, da alle Spezifikationen erfüllt sind. In dieser Aufgabe sollen Sie kein Nonogramm lösen, sondern stattdessen zu einem in `input` gegebenen Raster mit einigen schwarzen Zellen die passende Nonogramm-Spezifikation generieren.

Ist die übergebene 2-dimensionalen Liste leer (`Nil`), dann gibt es nichts (mehr) zu transponieren und das Ergebnis ist ebenfalls die leere Liste.

Im Rekursionsfall (unten) werden jeweils die ersten Elemente jeder Zeile extrahiert und der Rest rekursiv verarbeitet. Dabei tritt als zweiter Basisfall ein, dass alle Kopfelemente abgearbeitet sind und damit nur noch leere Listen in der Eingabe (also leere Zeilen in der Matrix) stehen. Auch in diesem Fall soll die leere Liste zurückgegeben werden.

Mittels `map` werden zunächst jeweils die Kopfelemente (`_.head`) aller Zeilen extrahiert. Für den rekursiven Aufruf von `transpose` benötigen genau das Gegenteil, nämlich alle Zeilen (`map`) _ohne_ ihr jeweiliges Kopfelement (`_.tail`).

<summary><i><span style='background:#b2ebf2'>Hinweis</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-info">

Die Liste wird rekursiv von vorne nach hinten abgearbeitet. Die ersten beiden `case` sind dabei die beiden möglichen Basisfälle.
1. `case`: Die Eingabe ist vollständig abgearbeitet (`Nil`) und es gilt `acc == 0`: D.h. die Eingabe endete mit einem weißen Feld (sonst wäre `acc > 0`) und damit kann hier die leere Liste `Nil` zurückgegeben werden.
2. `case`: Die Eingabe ist ebenfalls vollständig abgearbeitet (`Nil`), aber nun gilt offenbar `acc > 0`: D.h. die Eingabe endete mit einem schwarzen Feld, daher muss der Wert von `acc` in einer Liste zurückgegeben werden.
3. `case`: Die Eingabe ist nicht leer, das Kopfelement ist `White` und es gilt `acc == 0`: D.h. vor dem aktuellen weißen Feld muss es mind. noch ein weiteres gegeben haben (sonst wäre `acc > 0`). In diesem Fall rufen wir die Funktion rekursiv auf und übergeben ihr den neuen alten Wert `acc == 0` und die Restliste `rs`.
4. `case`: Die Eingabe ist nicht leer, das Kopfelement ist `White` und es gilt nun offensichtlich `acc > 0`: D.h. vor dem aktuellen weißen Feld muss es mind. ein schwarzes gegeben haben. Da das aktuelle Feld nun weiß ist, endet ein schwarzer Abschnitt. Daher muss der Wert von `acc` als neues Kopfelement derjenigen Liste vorangestellt werden, die aus der (vorgegebenen) Rekursion zurückgegeben wird.
5. `case`: Die Eingabe ist nicht leer, das Kopfelement ist `Black`, d.h. wir befinden uns in einem schwarzen Abschnitt. Daher erhöhen wir den Wert von `acc` um 1 und traversieren damit die Restliste rekursiv.

```scala
def describeHelper : Int => List[Cell] => List[Int] = acc => {
   case Nil if acc == 0 => Nil // Liste endet weiss
   case Nil => List(acc) // Liste endet schwarz
   case (White :: rs) if acc == 0 => 
        describeHelper(0)(rs)
   case (White :: rs) => acc :: describeHelper(0)(rs)
   case (Black :: rs) => describeHelper(acc + 1)(rs)
} 
```

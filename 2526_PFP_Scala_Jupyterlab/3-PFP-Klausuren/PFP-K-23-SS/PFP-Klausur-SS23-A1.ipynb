<div class="alert alert-block alert-info">

- `AnyVal` ist die Superklasse aller _Wert_-Typen (`Int`,`Char`,...), `AnyRef` die aller _Referenz_-Typen (`List`, Java-Klassen, ...) und `Any` ist der _Super-Typ_ dieser beiden und damit der höchste Datentyp in der Vererbungshierarchie.
- Mit `+T` wird die üblichere _"Kovarianz"_ und mit `-T` entsprechend die seltenere _"Kontravarianz"_ deklariert, wie im folgenden Beispiel:
  - `sealed abstract class Option[+T]`
  - `case object None extends Option[Nothing]`
  - `case class Some[+T](v: T) extends Option[T]`
  - Damit kann eine Funktion mit Rückgabetyp `Option[Int]` trotzdem den "Wert" `None` haben.
- Die Funktion _"MetaQuickSort"_ kann z.B. wie folgt definiert werden:  
  `def mqs: ((Int, Int) => Boolean) => List[Int] => List[Int] = ord => {`  
  `   case Nil => Nil`  
  `   case p :: ps => mqs(ord)(ps.filter(ord(_, p))) ::: p :: mqs(ord)(ps.filter(!ord(_, p)))`  
  `}`  
  Anschließend können zwei Funktionen davon abgeleitet werden, indem `mqs` _"unterversorgt"_ aufgerufen (also _"partiell"_ angewandt) wird:
  - `def sortAsc: List[Int] => List[Int] = mqs(_ < _)`
  - `def sortDesc: List[Int] => List[Int] = mqs(_ > _)`
- Da der Aufruf `c(0)` eine `LazyList` erzeugt, traversiert die `for`-comprehension diese ebenfalls _"faul"_ und erzeugt ihrerseits eine `LazyList`, die unter dem Bezeichner `v` unausgewertet bleibt.

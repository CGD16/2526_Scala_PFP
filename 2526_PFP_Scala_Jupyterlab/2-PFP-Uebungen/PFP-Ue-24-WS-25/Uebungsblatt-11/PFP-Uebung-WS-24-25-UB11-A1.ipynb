## Aufgabe 11.1: Seiteneffekte
Welche der folgenden _Java_-Methoden hat Seiteneffekte? Welche ist seiteneffektfrei?
```java
public class SeiteneffekteJava {
    private int a = 3;
    public static int b = 5;

    public void meth1(int x) {
        a = x;
    }

    public int meth2() {
        return a++;
    }

    public int meth3(int y) {
        return a + b + y;
    }

    public int meth4(int k) {
        b = b + 19 / b;
        return 15;
    }

    public int meth5(int k) {
        int a = ++k;
        return a;
    }

    public void meth6() {
        System.out.println("Hallo");
    }
}
```

<details>
<summary><i><span style='background:#b2ebf2'>Hinweis</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-info">

Eine _**Nebenwirkung**_ (oft auch: _**Seiteneffekt**_) bezeichnet allgemein jede Art einer _bleibenden Veränderung_ des Systemzustands infolge der Ausführung eines Code-Abschnitts.

Grundsätzlich ist damit **JEDE** beobachtbare Änderung gemeint:
- Werte der (Programm-)Variablen, aktueller Programm-Stack bzw. -Heap, Caches, ...
- Dateisystem oder Datenbanken (neue/geänderte/gelöschte Dateien bzw. Einträge), ...
- Peripheriegeräte (Bildschirmausgabe, Ausdruck auf Papier), Netzwerkkommunikation, ...
- Laufzeit (Dauer der Ausführung)
- ...

Meist (so auch hier in PFP) beschränkt man sich auf die "Variablenbelegungen" des Programms ("Programmzustand").

Ein Modul heißt _**werttreu**_, wenn es bei gleichen Eingaben (i.d.R. Parameter) stets die gleichen Ausgaben (meist Rückgabewert) ergibt, also _zustandslos_ ist.

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

- `meth1`: Nebenwirkung: verändert Attribut `a`! Werttreue irrelevant, weil `void`.
- `meth2`: Nebenwirkung: verändert Attribut `a`! Nicht werttreu: vom _Zustand_ von `a` abhängig!
- `meth3`: Keine Nebenwirkung: verändert keine Variablen. Nicht werttreu: vom _Zustand_ von `a` und `b` abhängig!
- `meth4`: Nebenwirkung: verändert Attribut `b`! Werttreu: Rückgabe ist konstant `15`.
- `meth5`: Keine Nebenwirkung: `a` und `k` sind nur lokale Variablen. Werttreu: nur von Eingabe `k` abhängig.
- `meth6`: Nebenwirkung: verändert Bildschirmzustand! Werttreue irrelevant, weil `void`.

</div>
</details>

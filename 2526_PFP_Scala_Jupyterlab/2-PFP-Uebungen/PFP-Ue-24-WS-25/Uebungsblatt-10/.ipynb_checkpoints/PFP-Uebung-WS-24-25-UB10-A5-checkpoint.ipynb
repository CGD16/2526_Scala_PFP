## Aufgabe 10.5: Rekursion, Mustervergleich, Listen
Implementieren Sie die nachfolgend genannten Funktionen in der Datei "Recursion.scala". Die Implementierung soll dabei unter Verwendung von Rekursion, Mustervergleich, der leeren Liste `Nil`, dem `::`-Operator zum Voranstellen eines einzelnen Elements an eine Liste von Elementen und selbst definierten Hilfsfunktionen erfolgen. Andere Listenfunktionen, wie z.B. `:::`, `++`, `:+`, `+:` und `length` sind nicht erlaubt.

**a)** `myLength: List[Any] => Int`: Berechnet die Länge der übergebenen Liste.

````scala
// Kleiner Testfall zu dieser Teilaufgabe:
assert(myLength(Nil) == 0); print("✅")
assert(myLength(List(1)) == 1); print("✅")
assert(myLength(List(1, 2)) == 2); print("✅")
assert(myLength(List(1, 2, 3, 4, 5, 6)) == 6); print("✅")
````

Prüfen Sie mittels _Pattern Matching_ (`case ... => ...`) welchen der folgenden Fälle die übergebene Liste erfüllt:
- Basisfall: Die Liste ist leer (`Nil`) => ihre Länge ist in diesem Fall `0`.
- Rekursionsfall: Die Liste besteht aus einem Kopfelement und einer Restliste `(kopf::restliste)` => ihre Länge ist `1 + Länge der Restliste`.

</div>

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

```scala
def myLength: List[Any] => Int = {
   case Nil => 0
   case _::ls => 1 + myLength(ls)
} 
```

**b)** `myConcat: (List[Any], List[Any]) => List[Any]`: Hängt die beiden übergebenen Listen aneinander.

````scala
// Kleiner Testfall zu dieser Teilaufgabe:
assert(myConcat(Nil,Nil) == Nil); print("✅")
assert(myConcat(List(1),Nil) == List(1)); print("✅")
assert(myConcat(Nil,List(9)) == List(9)); print("✅")
assert(myConcat(List(1),List(9)) == List(1,9)); print("✅")
assert(myConcat(List(1,2,3),List(7,8,9)) == List(1,2,3,7,8,9)); print("✅")
````

<details>
<summary><i><span style='background:#b2ebf2'>Hinweis</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-info">

Sei $\{a,b,c,d,e\} ^\frown \{1,2,3,4,5\} \mapsto \{a,b,c,d,e,1,2,3,4,5\}$ die _Konkatenation_ (Aneinanderhängen) der beiden Folgen und $a :: \{b,c,d,e\} \mapsto \{a,b,c,d,e\}$ das Voranstellen von $a$.
Dann können wir zwei Fälle unterscheiden:
- $\{\} ^\frown \{1,2,3,4,5\} \mapsto \{1,2,3,4,5\}$
- $\{a,b,c,d,e\} ^\frown \{1,2,3,4,5\} \mapsto a :: (\{b,c,d,e\} ^\frown \{1,2,3,4,5\})$

Prüfen Sie mittels _Pattern Matching_ (`case ... => ...`) welchen der folgenden Fälle das übergebene Tupel aus zwei Listen `(as,bs)` erfüllt:
- Basisfall: Die erste Liste `as` ist leer (`Nil`) => das Ergebnis ist die zweite Liste `bs`.
- Rekursionsfall: Die erste Liste `as` besteht aus einem Kopfelement `x` und einer Restliste `xs`, die Eingabe ist also `(x::xs, bs)` => für das Ergebnis muss das Kopfelement `x` dem Ergebnis des rekursiven Aufrufs der Funktion mit der Restliste `xs` und `bs` vorangestellt werden.

</div>
</details>

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

```scala
def myConcat: (List[Any], List[Any]) => List[Any] = {
   case (Nil, ys) => ys
   case (x::xs, ys) => x::myConcat(xs, ys)
} 
```

</div>
</details>

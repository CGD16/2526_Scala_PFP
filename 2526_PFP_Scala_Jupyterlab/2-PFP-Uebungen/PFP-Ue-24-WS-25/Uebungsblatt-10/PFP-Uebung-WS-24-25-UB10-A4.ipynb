## Aufgabe 10.4: Funktionen – Vertiefung
Erstellen Sie die Datei "Functions.scala" und implementieren Sie darin die folgenden Funktionen:

**a)** `abs: Int => Int`: Berechnet den Absolutwert des übergebenen Wertes. Wenden Sie hierbei das Konzept des Mustervergleichs an.

````scala
// Kleiner Testfall zu dieser Teilaufgabe:
assert(List(-5,-1,0,1,5).map(abs) == List(5,1,0,1,5)); print("✅")
````

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

```scala
def abs: Int => Int = {
   case x if x >= 0 => x
   case x => -x
} 
```

</div>
</details>

**b)** `fac: Int => Int`: Berechnet die Fakultät des übergebenen Wertes.

````scala
// Kleiner Testfall zu dieser Teilaufgabe:
assert(List(0,1,2,3,4,5).map(fac) == List(1,1,2*1,3*2*1,4*3*2*1,5*4*3*2*1)); print("✅")
````

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

```scala
def fac: Int => Int = {
   case 0 => 1
   case n if n >= 0 => n * fac(n - 1)
} 
```

</div>
</details>

**c)** `quadraticEquation: (Double, Double, Double) => List[Double]`: Berechnet die Liste reeller Ergebnisse der quadratischen Gleichung _ax² + bx + c = 0_. Die Funktion wird in der Form `quadraticEquation(a, b, c)` aufgerufen. Zum Berechnen der Quadratwurzel dient die vordefinierte Funktion `scala.math.sqrt: Double => Double`.

````scala
// Kleiner Testfall zu dieser Teilaufgabe:
assert(quadraticEquation(2,3,4) == Nil); print("✅")
assert(quadraticEquation(4,-4,1) == List(0.5)); print("✅")
assert(quadraticEquation(3,-6,-9) == List(-1,3) | quadraticEquation(3,-6,-9) == List(3,-1)); print("✅")
````

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

```scala
def quadraticEquation: (Double, Double, Double) => List[Double] = (a, b, c) => {
   b * b - 4 * a * c match {
      case d if(d < 0) => Nil
      case 0 => List(-b / (2 * a))
      case d => List((-b + scala.math.sqrt(d)) / (2 * a), (-b - scala.math.sqrt(d)) / (2 * a))
   }
} 
```

</div>
</details>

**d)** Definieren Sie eine Funktion `f3`, die als Parameter ein Tupel `(x, y)` entgegennimmt und das Tupel `(y, x)` zurückliefert. Was ist der Unterschied zwischen `f2` und `f3`?

````scala
// Kleiner Testfall zu dieser Teilaufgabe:
assert(f3(42,666) == (666,42)); print("✅")
````

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

```scala
def f3: (Int,Int) => (Int,Int) = (x,y) => (y,x)
def f3generic[T,R]: (T,R) => (R,T) = (x,y) => (y,x)
```

</br>

Unterschied zwischen `f2` und `f3`:
  - `f2` erwartet "zwei Parameterlisten" mit "je einem Parameter". Die "Parameter" können auch einzeln und unabhängig übergeben werden (Currying).
  - `f3` erwartet "eine Parameterliste" mit "einem Parameter" vom Tupel-Typ.

</div>
</details>

**f)** Rufen Sie die Funktionen `Beispiel.g` und `Beispiel.h` mit den Werten `1`, `2`, `3`, `4`, `10` und `-20` auf. Was berechnen diese beiden Funktionen?

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

```scala
Beispiel.g(1)
Beispiel.g(2)
Beispiel.g(3)
Beispiel.g(4)
Beispiel.g(10)
Beispiel.g(-20)
//
for (x <- List(1,2,3,4,10,-20)) yield Beispiel.h(x)
```

</br>

Was berechnen die Funktionen?
  - `g` berechnet für negative Argumente x ≤ 0 die Abbildung x ↦ x² + 3, sonst deren Kubikzahl x ↦ x³.
  - `h` berechnet für Argumente x ≥ 1 die x-te Fibonacci-Zahl (1-indiziert bei 1 beginnend), sonst gibt sie ihr Argument zurück.

</div>

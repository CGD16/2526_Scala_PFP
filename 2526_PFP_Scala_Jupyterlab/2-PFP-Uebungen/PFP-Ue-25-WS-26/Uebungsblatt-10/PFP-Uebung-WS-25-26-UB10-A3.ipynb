## Aufgabe 10.3: Mustervergleich in Scala

**a)** Was versteht man unter Mustervergleich in Programmiersprachen?

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

_Mustervergleich_ (engl. _Pattern Matching_) ist eine Erweiterung der mehrfachen Fallunterscheidung.

Dabei werden z.B. konkrete Werte, komplexe Datenstrukturen (bzw. Objekte) oder sogar Datentypen strukturell mit den vorgegebenen "Mustern" in der vorgegebenen Reihenfolge ("von oben nach unten") verglichen, bis ein "Muster" auf die Eingabe "passt" - das erst-beste gewinnt!

Dabei können in einem "Muster" auch Variablen oder Prädikate vorkommen: Erstere werden passend belegt, letztere überprüft und verhindern ggf. dass das "Muster passt".

</div>
</details>

**b)** Welche Ergebnisse liefern die Funktionen `f` und `g` mit den Parametern `Nil`, `List()`, `19`, `List(4, 5, 6)`, `List.range(10, 21)`, `List(3)` und `"Hallo Welt"`?

```scala
def f: List[Int] => Int = {
   case h::g::ts => h * 5 * g
   case _ => -1
}

def g: List[Int] => List[Int] = {
   case Nil => List(15)
   case _::ts => List(ts.length)
   case _ => List(20)
}
```

```scala
f(Nil)
```

```scala
f(List())
```

```scala
f(19)
```

```scala
f(List(4, 5, 6))
```

```scala
f(List.range(10, 21))
```

```scala
f(List(3))
```

```scala
f("Hallo Welt")
```

```scala
g(Nil)
```

```scala
g(List())
```

```scala
g(19)
```

```scala
g(List(4, 5, 6))
```

```scala
g(List.range(10, 21))
```

```scala
g(List(3))
```

```scala
g("Hallo Welt")
```

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

- `f(Nil)` => `-1` // der erste `case` trifft nur zu, wenn _mindestens zwei_ Elemente in der Liste sind - `Nil` ist aber die leere Liste
- `f(List())` => `-1` // identisch zu `Nil` (oben)
- `f(19)` => `Type Mismatch Error!` // `f` erwartet eine ganze Liste mit `Int`-Werten
- `f(List(4, 5, 6))` => `100` // `h * 5 * g` == `4 * 5 * 5`, weil `h==4` und `g==5` sind
- `f(List.range(10, 21))` => `550` // `h * 5 * g` == `10 * 5 * 11`
- `f(List(3))` => `-1` // identisch zu `Nil` (oben)
- `f("Hallo Welt")` => `Type Mismatch Error!` // `f` erwartet eine ganze Liste mit `Int`-Werten

---

- `g(Nil)` => `List(15)` // der erste `case` trifft zu
- `g(List())` => `List(15)` // identisch zu `Nil` (oben) - `Nil` ist die leere Liste
- `g(19)` => `Type Mismatch Error!` // `g` erwartet eine ganze Liste mit `Int`-Werten
- `g(List(4, 5, 6))` => `List(2)` // der zweite `case` spaltet die Liste in "`_=4 :: ts=List(5, 6)`" => das Ergebnis ist daher `List(List(5, 6).length) == 2`
- `g(List.range(10, 21))` => `List(10)` // analog zum vorangehenden Fall => `List(List(11, 12, ..., 20).length)`
- `g(List(3))` => `List(0)` // analog zum vorangehenden Fall "`_=3 :: ts=Nil`" => `List(Nil.length)`
- `g("Hallo Welt")` => `Type Mismatch Error!` // `g` erwartet eine ganze Liste mit `Int`-Werten

</div>
</details>

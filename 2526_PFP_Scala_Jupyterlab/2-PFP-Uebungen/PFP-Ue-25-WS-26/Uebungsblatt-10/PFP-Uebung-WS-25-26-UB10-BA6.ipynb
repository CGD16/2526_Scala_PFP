## <u>Bonus</u>aufgabe 10.6: Texte ordentlich formatieren
Gemeinsam mit einigen Kommilitoninnen und Kommilitonen haben Sie eine Vorlesungszusammenfassung für die PFP-Vorlesung geschrieben. Doch durch die vielen Mithelfenden ist der Text sehr inkonsistent formatiert. Sie beschließen, Ihre neu gewonnenen Scala-Kenntnisse zu nutzen, um die Formatierung zu vereinheitlichen. Implementieren Sie daher die folgenden Funktionen in der Datei "TextCleaner.scala".  
**Wichtig:** Verwenden Sie nur Mustervergleiche und Rekursion!

**a)** `contains: (List[Char], Char) => Boolean`: Als Einstieg schreiben Sie sich eine Hilfsfunktion `contains(cs, c)`, die überprüft, ob ein Zeichen `c` in einer Liste von Zeichenketten `cs` enthalten ist.

```scala
def contains: (List[Char], Char) => Boolean = {
    case (Nil, c) => false
    case (l::cs, c) if c == l => true
    case (l::cs, c) => false || contains(cs, c)
}
```

```scala
// Kleiner Testfall zu dieser Teilaufgabe:
{val f1: (List[Char], Char) => Boolean = contains}
assert(contains("abcdefg".toList, 'c')); print("✅")
```

**b)** `clean: (List[Char], List[Char]) => List[Char]`: Die Funktion `clean(cs, ps)` erhält eine Liste von Zeichen `cs` sowie eine Liste von Satzzeichen `ps` und soll eine "bereinigte" Form von `cs` zurückgeben. Dazu wendet sie folgende Regeln an:
  - Mehrfache Leerzeichen sollen zu einem Leerzeichen verschmolzen werden.
  - Leerzeichen vor Satzzeichen (alle Zeichen aus `ps`) sollen entfernt werden (Tipp: Prüfen Sie mit der vorher definierten Funktion `contains`, ob ein Zeichen ein Satzzeichen ist).

Beispiel:  
`clean("Guten␣␣Tag␣!".toList, List('!'))`  
` == List('G','u','t','e','n','␣','T','a','g','!')`

```scala
def clean: (List[Char], List[Char]) => List[Char] = {
    case (Nil, ps) => Nil
    case (c1::Nil, ps) if c1 == ' ' => Nil
    case (c1::Nil, ps) => List(c1)
    case (c1::c2::cs, ps) if c1 == ' ' && c2 == ' ' => clean(c2::cs, ps)
    case (c1::c2::cs, ps) if c1 == ' ' && contains(ps, c2) => clean(c2::cs, ps)
    case (c::cs, ps) => c :: clean(cs, ps)
}
```

```scala
// Kleiner Testfall zu dieser Teilaufgabe:
{val f2: (List[Char], List[Char]) => List[Char] = clean}
assert(clean("Guten  Tag !".toList, List('!')) == "Guten Tag!".toList); print("✅")
```

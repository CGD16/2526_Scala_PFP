**b)** Welches Ergebnis liefern die Funktionen `head` und `tail` der Liste `List(4, 2, 8, 5, 8)`?

````scala
List(4, 2, 8, 5, 8).head == 4
List(4, 2, 8, 5, 8).tail == List(2, 8, 5, 8)
````

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

- `List(4, 2, 8, 5, 8).head`
  - Ergebnis: `4`
  - Erklärung: `head` liefert den Kopf einer (nicht-leeren!) Liste, also das erste Element
- `List(4, 2, 8, 5, 8).tail`
  - Ergebnis: `List(2, 8, 5, 8)`
  - Erklärung: `tail` liefert die Restliste ohne das Kopfelement

</div>
</details>

**c)** Welche Ergebnisse liefern die folgenden Ausdrücke?

````scala
List(5, 3, 6, 7, 8)(3)
````

````scala
List(5, 3, 6, 7, 8)(3) == 7
````

````scala
List(5, 3, 6, 7, 8):::(3::(5::7::78::22::Nil))
````

````scala
List(5, 3, 6, 7, 8):::(3::(5::7::78::22::Nil)) == List(5, 3, 6, 7, 8, 3, 5, 7, 78, 22)
````

````scala
List.range(3, 11).length
````

````scala
List.range(3, 11).length == 8
````

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

- `List(5, 3, 6, 7, 8)(3)`
  - Ergebnis: `7`
  - Erklärung: der Zugriff über den Index ruft implizit die `List`-Funktion `def apply(n: Int): A` auf, die das `n`-te Element (`0`-indiziert!) zurückgibt
- `List(5, 3, 6, 7, 8):::(3::(5::7::78::22::Nil))`
  - Ergebnis: `List(5, 3, 6, 7, 8, 3, 5, 7, 78, 22)`
  - Erklärung: `:::` konkateniert die beiden Listen `List(5, 3, 6, 7, 8)` und `(3::(5::7::78::22::Nil))` (d.h. "hängt sie aneinander")
- `List.range(3, 11).length`
  - Ergebnis: `8`
  - Erklärung: `List.range(3, 11)` erzeugt `List(3, 4, 5, 6, 7, 8, 9, 10)` mit `8` Elementen

</div>
</details>

## <u>Bonus</u>aufgabe 12.4: Schriftliches Wurzelziehen
In dieser Aufgabe setzen Sie einen Algorithmus zum [Ziehen der quadratischen Wurzel](https://de.wikipedia.org/wiki/Schriftliches_Wurzelziehen) mit beliebiger Genauigkeit in Scala um. Die Berechnung von $\sqrt{654{,}321}$ beispielsweise funktioniert so:
- **S1** Unterteile die Zahl vom Dezimaltrennzeichen aus in beide Richtungen in Gruppen zu je $2$ Ziffern (die erste Gruppe ist ggf. kleiner und die letzte wird mit $0$ aufgefüllt): $\sqrt{6|54,32|10|00|\ldots}$
- **S2** Die erste Ziffer der Quadratwurzel ist die größte Zahl $a$, deren Quadrat gerade noch kleiner oder gleich der ersten Gruppe ist: Da $2^2 \leq 6 < 3^2$ ist, lautet die erste Stelle also $a = 2$.
- **S3** Das Quadrat dieser ersten Ziffer wird von der ersten Gruppe abgezogen und an das Ergebnis wird die nächste Gruppe $\underline{54}$ angehängt: $b := (6 - a^2) \cdot 100 + \underline{54} = 2\underline{54}$
- **S4** Nun wird die größte Ziffer $c$ gesucht, für die $(a \cdot 20 + c) \cdot c \leq b$ gerade noch gilt: Wegen $(2 \cdot 20 + 5) \cdot 5 \leq 254 < (2 \cdot 20 + 6) \cdot 6$ lautet die zweite Stelle der Wurzel daher $c = 5$.
- **S5** Die Zwischenergebnisse $b$ und $a$ werden mit $d := (a \cdot 20 + c) \cdot c$ aus **S4** und der nächsten Gruppe $\underline{32}$ aktualisiert: $b \leftarrow (b - d) \cdot 100 + \underline{32} = 29\underline{32}$ und $a \leftarrow (a \cdot 10 + c) = 25$.
- **S6** Anschließend wiederholt man die Schritte **S4** und **S5** _ad infinitum_, wobei man nach der letzten Gruppe "beliebig" lange mit $\underline{00}$ auffüllt. Sobald die nächste Gruppe unmittelbar "hinter" dem Dezimalpunkt steht, wird auch in der Lösung der Dezimalpunkt gesetzt.

Der weitere Ablauf ab $a = 25$, $b = 29\underline{32}$ und $\sqrt{6|54,\underline{32}|10} = 25,\ldots$ lautet also:  
$\blacktriangleright$ $(25 \cdot 20 + 5) \cdot 5 \leq 29\underline{32} < (25 \cdot 20 + 6) \cdot 6$ $\rightsquigarrow$ $c = 5$  
$\vartriangleright$ $b \leftarrow 407\underline{10}$ sowie $a \leftarrow 255$ <span style="display:inline-block;width:265px;"></span> $\sqrt{6|54,32|\underline{10}} = 25,5\ldots$  
$\blacktriangleright$ $(255 \cdot 20 + 7) \cdot 7 \leq 407\underline{10} < (255 \cdot 20 + 8) \cdot 8$ $\rightsquigarrow$ $c = 7$  
$\vartriangleright$ $b \leftarrow 4961\underline{00}$ sowie $a \leftarrow 2557$ <span style="display:inline-block;width:250px;"></span> $\sqrt{6|54,32|10} = 25,57\ldots$  
$\blacktriangleright$ $(2557 \cdot 20 + 9) \cdot 9 \leq 4961\underline{00} < (2557 \cdot 20 + 10) \cdot 10$ $\rightsquigarrow$ $c = 9$  
$\vartriangleright$ $b \leftarrow 35759\underline{00}$ sowie $a \leftarrow 25579$ <span style="display:inline-block;width:235px;"></span> $\sqrt{6|54,32|10} = 25,579\ldots$

$\ldots$

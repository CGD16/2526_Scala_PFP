**b)** Nennen Sie die Ihnen bekannten Ausprägungen innerhalb dieser Paradigmen und ordnen Sie diesen die Ihnen bekannten Programmiersprachen zu.

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

- _imperativ_: Der Programmierer gibt den Programmablauf explizit vor und bestimmt damit, **wie** ein Ergebnis ermittelt werden soll.
  - prozedural (C, Pascal)
  - objekt-orientiert (C++, Java, C#)

<hr>

- _deklarativ_: Der Programmierer beschreibt lediglich die Ergebnismenge und gibt auf diese Weise an, **was** ermittelt werden soll.
  - funktional (Haskell)
  - logisch (Prolog)

</div>
</details>

**c)** Welchem Paradigma und welcher Ausprägung lässt sich _Scala_ zuordnen?

<details>
<summary><i><span style='background:#c8e6c9'>Lösungsvorschlag</span> (anzeigen/verbergen)</i></summary>
<div class="alert alert-block alert-success">

_Scala_ ist eine _Multi-Paradigmen-Sprache_ und kombiniert _objekt-orientierte_ und _funktionale_ Programmierung.

_Scala_ bietet sowohl _imperative_ als auch _deklarative_ Lösungswege.

</div>
</details>
